ANALYSIS

```python
#Total Income and Expense

total_income = df[df["TYPE"] == "Income"]["AMOUNT"].sum()
total_expense = df[df["TYPE"] == "Expense"]["AMOUNT"].sum()
net_balance = total_income - total_expense

print("Total Income:", f"{total_income:.2f}")
print("Total Expense:", f"{total_expense:.2f}")
print("Net Balance:", f"{net_balance:.2f}")
```

```python
#Time Analysis

df["DATE"] = pd.to_datetime(df["DATE"])  #date format

monthly_summary = df.groupby([df["DATE"].dt.to_period("M"), "TYPE"])["AMOUNT"].sum().unstack()
print(monthly_summary)
```

```python
#Expense by Category

expense_by_category = df[df["TYPE"] == "Expense"].groupby("CATEGORY")["AMOUNT"].sum().sort_values(ascending=False)

print(expense_by_category)
```

```python
#Income by Category

income_by_category = df[df["TYPE"] == "Income"].groupby("CATEGORY")["AMOUNT"].sum().sort_values(ascending=False)

print(income_by_category)
```

```python
#Account Analysis

account_summary = df.groupby(["ACCOUNT", "TYPE"])["AMOUNT"].sum().unstack().fillna(0)

print(account_summary)
```

VISUALIZATION

```python
#Monthly Summary Line Chart

plt.figure(figsize=(10,6))
monthly_summary.plot(kind="line", marker="o")
plt.title("Monthly Income vs Expense")
plt.ylabel("Amount")
plt.xlabel("Month")
plt.legend(title="Type")
plt.grid(True)
plt.show()
```

```python
# Expense by Category Chart

# Keep only Top 5
top_expenses = expense_by_category.sort_values(ascending=False).head(5)

plt.figure(figsize=(8,8))
top_expenses.plot(kind="pie", autopct="%1.1f%%", startangle=140, cmap="tab20")
plt.title("Top 5 Expense Categories")
plt.ylabel("")
plt.show()
```

```python
# Income by Category Chart

# Top 5 only
top_income = income_by_category.sort_values(ascending=False).head(5)

plt.figure(figsize=(7,7))
top_income.plot(kind="pie", autopct="%1.1f%%", startangle=140, cmap="tab20")
plt.title("Top 5 Income Categories")
plt.ylabel("")
plt.show()
```
